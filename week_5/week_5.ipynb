```python
def my_f_3(im_100):
    im_1=im_100
    m,n,p=im_1.shape
    im_2=np.zeros((m,n,3),dtype=int)
    for m in range(im_1.shape[0]):
        for n in range(im_1.shape[1]):
            im_2[m,n,0]=int(im_1[m,n,0])
            im_2[m,n,1]=int(im_1[m,n,0])
            im_2[m,n,2]=int(im_1[m,n,0])
    return im_2 
```

```python
im_2= my_f_3(im_1)
```

```python
plt.imshow( im_2)
plt.show()
```

```python
def my_f_4(im_500 ,zoomX):
    m,n,p=im_500.shape
    new_m=int(m/zoomX)
    new_n=int(n/zoomX)
    new_p=int(p/zoomX)
    im_600=np.zeros((new_m,new_n,3),dtype=int)
    for m in range(new_m):
        for n in range(new_n):
            im_600[m,n,0]=int(im_500[m*2,n*2,0])
            im_600[m,n,1]=int(im_500[m*2,n*2,1])
            im_600[m,n,2]=int(im_500[m*2,n*2,2])
    return im_600 

```

```python
 
plt.imshow( my_f_3(255-im_2))
plt.show()
```
