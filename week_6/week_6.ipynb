```python
im_1=plt.imread('kaneki_kun.jpg')
plt.imshow(im_1)
plt.show()
```

```python
 
im_2=np.zeros((888,750),dtype=np.uint8)
```

```python
im_2=im_1[:,:,0] # siyah-beyaz resim
im_3=im_1[:,:,0] # filtreleme yapılacak resim
```

```python
im_3=np.zeros((888,750),dtype=np.uint8)
```

```python
plt.imshow(im_2)
plt.show()
```

```python
plt.imshow(im_2,cmap='gray')
plt.show()
```

```python
#smoothing for 1 time  
m,n=im_2.shape
for i in range(1,m-1):
    for j in range(1,n-1):
        s= \
        im_2[i-1,j-1]/9+ \
        im_2[i-1,j]/9+ \
        im_2[i-1,j+1]/9+ \
        im_2[i,j-1]/9+ \
        im_2[i,j]/9+ \
        im_2[i,j+1]/9+ \
        im_2[i+1,j-1]/9+ \
        im_2[i+1,j]/9+ \
        im_2[i+1,j+1]/9 
        s=int(s)
        # print(s ,end=' * ')
        im_3[i,j]=s
```

```python
#smoothing for 2nd time on the same photo
im_4=np.zeros((888,750),dtype=np.uint8)
m,n=im_3.shape
for i in range(1,m-1):
    for j in range(1,n-1):
        s= \
        im_3[i-1,j-1]/9+ \
        im_3[i-1,j]/9+ \
        im_3[i-1,j+1]/9+ \
        im_3[i,j-1]/9+ \
        im_3[i,j]/9+ \
        im_3[i,j+1]/9+ \
        im_3[i+1,j-1]/9+ \
        im_3[i+1,j]/9+ \
        im_3[i+1,j+1]/9 
        s=int(s)
        # print(s ,end=' * ')
        im_4[i,j]=s
```

```python

plt.subplot(1,2,1) ,plt.imshow(im_2,cmap='gray')#smooth
plt.suptitle('smoothing', fontsize=30)
plt.subplot(1,2,2),plt.imshow(im_4 ,cmap='gray')# more smooth
 
plt.show()
```
