```python
m,n,p = im_1.shape    #return the values of shape function to other varibles that m =x ,n=y ,p =color type RGB
```

```python
m,n,p
```

```python
#white and grey photo 
#the first make white photo with the same dimension of our photo
new_im = np.zeros((m,n),dtype =float )
```

```python
#copying the pixel values of the orginal photo to the white photo pixel by pixel 
for i in range (m):
    for j in range (n):
        
        #here we take the values of colors Red,Green,Blue and mix them to return the color of that  [i,j] pixel
        s=(im_1[i,j,0]+im_1[i,j,1]+im_1[i,j,2])/3  
        
        #throwing the returned value of S  to the [i,j] pixel 
        new_im[i,j]=s
```

```python
#after copying we can change the color without damaging the orginal photo 
plt.imshow(new_im,cmap='gray')
plt.show() 
```

```python
#saving the made photo with gray color that without showing us how look like is 
plt.imsave('WhiteBlackImage.png',new_im,cmap='gray')
```

```python
#return the photo by change the x position in y and y in x  that give us 90 degree left side 
 
returned_im = np.zeros((m,n),dtype =float )
for i in range(m):
    for j in range (n):
        s=im_1[i,j,0]+im_1[i,j,1]+im_1[i,j,2]/3
        returned_im[j,i]=s
```

```python
plt.imshow(returned_im)
plt.show() 

```

```python
#saving as new image as gray 
plt.imsave('leftSide.png',returned_im,cmap='gray')
```

```python
#return the photo by change the x position in y and y in x  that give us 90  degree for right side 
returned_im=np.zeros((m,n),dtype = float )
for i in range(m):
    for j in range (n):
        s=im_1[i,j,0]+im_1[i,j,1]+im_1[i,j,2]/3
        returned_im[-j,-i]=s
plt.imshow(returned_im)
plt.show()         
#saving as new image as gray 
plt.imsave('rightSide.png',returned_im,cmap='gray')
```

```python
#return the photo by change the x position in y and y in x  that give us updown  degree for right side 
returned_im=np.zeros((m,n),dtype = float )
for i in range(m):
    for j in range (n):
        s=im_1[i,j,0]+im_1[i,j,1]+im_1[i,j,2]/3
        #all the changing from here change the dimensions 
        returned_im[-i,-j]=s
plt.imshow(returned_im)
plt.show()  
#saving as new image as gray 
plt.imsave('updown.png',returned_im,cmap='gray')
```
