```python
def play_with_photo_color(im_3,r=5,g=5 ,b =5):
    
    im_1=im_3
    m,n,p=im_1.shape
    im_2=np.zeros((m,n,3),dtype=int)
 
    for m in range(im_1.shape[0]):  
        for n in range(im_1.shape[1]):
            im_2[m,n,0]=im_1[m,n,0]-r # this parameter change the color of red `r`
            im_2[m,n,1]=im_1[m,n,1]-g  # for green color `g` 
            im_2[m,n,2]=im_1[m,n,2]-b   # for blue color `b`
      
    return im_3
im_ret=play_with_photo_color(im_1,-50 ,10,220)
plt.imshow(im_ret)
plt.show()
```

```python
def ZoomIn(im_4 ,zoomX2):
    #gnderilen mxn lik resmi m/2 x n/2 lik yapan fonksiyon
    x,y,p=im_4.shape
    new_x=int(m/zoomX2)
    new_y=int(n/zoomX2)
    im_5=np.zeros((new_x,new_y),dtype=int)
    for x in range(new_x): 
        for y in range(new_y):
            s=(im_4[x*2,y*2,0]+im_4[x*2,y*2,1]+im_4[x*2,y*2,2])/3
            im_5[x,y]=int(s)
    return im_5
```

```python
def quality_control(im_7):
    x,y,p=im_7.shape
    new_x=int(x/2)
    new_y=int(y/2)
    im_8=np.zeros((new_x,new_y,3),dtype=int)
    for x in range(new_x): 
        for y in range(new_y):
            im_8[x,y,0]=int(im_7[x*2,y*2,0])
            im_8[x,y,1]=int(im_7[x*2,y*2,1])
            im_8[x,y,2]=int(im_7[x*2,y*2,2])
    return im_8

plt.imshow(im_1)
plt.show()
im_8=quality_control(im_1)
plt.imshow(im_8)
plt.show()
im_9=quality_control(im_8)
plt.imshow(im_9)
plt.show()
im_10=quality_control(im_9)
plt.imshow(im_10)
plt.show()
```
